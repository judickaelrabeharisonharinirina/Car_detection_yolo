```python
import os

# 1. Définition des chemins
dataset_dir = "VehiclesDetectionDataset"
yaml_path = os.path.join(dataset_dir, "dataset.yaml")

print("verification")

# 2. Vérification des dossiers indispensables pour YOLO
dossiers_requis = ['train/images', 'train/labels', 'valid/images', 'valid/labels', 'test/images', 'test/labels']
tout_est_ok = True

for d in dossiers_requis:
    chemin_complet = os.path.join(dataset_dir, d)
    if os.path.exists(chemin_complet):
        print(f"Dossier trouvé : {d} ({len(os.listdir(chemin_complet))} fichiers)")
    else:
        print(f"Dossier MANQUANT : {d}")
        tout_est_ok = False

# 3. Lecture et affichage du fichier data.yaml
if os.path.exists(yaml_path):
    with open(yaml_path, "r") as f:
        print(f.read())
else:
    print(f" Le fichier {yaml_path} n'est pas trouver ")
    tout_est_ok = False

if tout_est_ok:
    print("\nLance")
```
