```python
from ultralytics import YOLO
import os

yaml_path = "VehiclesDetectionDataset/dataset.yaml"

if __name__ == '__main__':
    if os.path.exists(yaml_path):
        print("🚀 Lancement de l'entraînement sur la GTX 1080...")
        
        model = YOLO("yolov8n.pt") 
        
        model.train(
            data=yaml_path,
            epochs=50,
            imgsz=640,
            batch=16,          # Avec 8Go de VRAM, batch=16 passera sans problème
            device=0,          # <--- 0 signifie "première carte graphique NVIDIA"
            name="Vehicule_detection"
        )
        
        print("\n🎉 Entraînement terminé avec succès sur GPU !")
    else:
        print(f"❌ Erreur : Le fichier '{yaml_path}' est introuvable.")
```
