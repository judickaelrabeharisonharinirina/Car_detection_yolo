```python
import cv2
from ultralytics import YOLO

def main():
    # 1. Charger le modèle entraîné (ajuste le chemin si nécessaire)
    model_path = "runs/detect/Vehicule_detection/weights/best.pt"
    
    print(f"Chargement du modèle : {model_path}")
    model = YOLO(model_path)

    # 2. Ouvrir le flux de la webcam (0 = webcam intégrée ou principale)
    print("🎥 Initialisation de la webcam...")
    cap = cv2.VideoCapture(0)

    # Optionnel : Forcer une résolution fluide si ta webcam le gère (ex: 1280x720)
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

    if not cap.isOpened():
        print("❌ Impossible d'accéder à la webcam.")
        return

    print("🚀 Tracking en direct lancé ! Montre des objets/véhicules à la caméra.")
    print("👉 Appuie sur 'q' pour quitter.")

    # 3. Boucle de capture en temps réel
    while cap.isOpened():
        success, frame = cap.read()
        
        if success:
            # On lance le tracking sur le GPU (device=0)
            # stream=True est ajouté ici : cela optimise la gestion de la mémoire pour les flux en direct
            # conf=0.60 signifie qu'il doit être sûr à 60% minimum pour afficher la boîte
            results = model.track(frame, persist=True, tracker="bytetrack.yaml", device=0, stream=True, conf=0.70)
            
            # Récupérer le premier résultat du générateur stream=True
            for r in results:
                annotated_frame = r.plot()
            
            # Afficher le flux vidéo en direct
            cv2.imshow("YOLOv8 Real-Time Tracking (GTX 1080)", annotated_frame)
            
            # Quitter proprement avec 'q' ou en fermant la fenêtre
            if cv2.getWindowProperty("YOLOv8 Real-Time Tracking (GTX 1080)", cv2.WND_PROP_VISIBLE) < 1:
                break
            if cv2.waitKey(1) & 0xFF == ord("q"):
                break
        else:
            print("Impossible de lire le flux de la webcam.")
            break

    # Libération des ressources
    cap.release()
    cv2.destroyAllWindows()
    print("Webcam coupée.")

if __name__ == '__main__':
    main()
```
